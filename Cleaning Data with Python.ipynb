```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#Getting Data
data = pd.read_excel("Table of Size Standards_Effective July 14 2022.xlsx", sheet_name='table_of_size_standards-all')
data
```

```python
#selecting only the relevant records
data = data [data['NAICS Codes'].isnull()]
```

```python
data.info()
```

```python
data.head()
```

```python
data['LookupCodes'] = data['NAICS Industry Description'].str.slice(start =7, stop =9) # Slicing NAICS Indus Desp Section 11 create new column name is LookupCodes
```

```python
data.head()
```

```python
data['NAICS Industry Description'].str.len()
```

```python
data['Sector'] = data['NAICS Industry Description'].str.slice(start =12, stop =538) # data['NAICS Industry Description'].str.len() insert stop 537 place NaN accured. 538 is hight number of length
```

```python
data.head()
```

```python
# [[ ]] using For df[[colname(s)]], the interior brackets are for list, and the outside brackets are indexing operator,
#i.e. you must use double brackets if you select two or more columns. With one column name, single pair of brackets returns
#a Series, while double brackets return a dataframe.

data = data[['NAICS Industry Description', 'LookupCodes', 'Sector']] 
data
```

```python
data.info()
```

```python
data.head()
```

```python
 # Removed for NA items
#Ex: 615	(These NAICS codes shall not be used to classi...NA	codes shall not be used to classify Governmen...data
data[data['LookupCodes'].str.isnumeric()]
```

```python

data = data[data['LookupCodes'].str.isnumeric()]
```

```python
data
```

```python
# Replacing some values in sector
# using Pandas
data.loc[data['Sector'] == '33 – Manufacturing', 'Sector'] = 'Manufacturing'
```

```python
# Replacing some values in sector
# using Numpy
data['Sector'] = np.where(data['Sector'] == '45 – Retail Trade', 'Retail Trade', data['Sector'])
data['Sector'] = np.where(data['Sector'] == '49 – Transportation and Warehousing','Transportation and Warehousing', data['Sector'])
```

```python
data
```

```python
#Type 1 create new data
data.loc[len(data.index)] = ['Sector 31 – 33 – Manufacturing', 32, 'Manufacturing']
data.loc[len(data.index)] = ['Sector 31 – 33 – Manufacturing', 33, 'Manufacturing']
#Type 2 create new data
df2 = {'NAICS Industry Description': 'Sector 44 - 45 – Retail Trade', 'LookupCodes': 45, 'Sector':'Retail Trade'}
df2 = {'NAICS Industry Description': 'Sector 48 - 49 – Transportation and Warehousing', 'LookupCodes': 49, 'Sector':'Transportation and Warehousing'}
data = data.append(df2, ignore_index = True)
data
```
